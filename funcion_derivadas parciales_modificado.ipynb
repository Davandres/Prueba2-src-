```python
def der_parcial_2_2(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 2 (a_2). La ecuación es de la forma:
    c_2 * a_2 + c_1 * a_1 + c_0 * a_0 = c_ind.

    ## Parameters

    ``xs``: lista de valores de x.

    ``ys``: lista de valores de y.

    ## Return

    ``c_2``: coeficiente del parámetro 2 (a_2).

    ``c_1``: coeficiente del parámetro 1 (a_1).

    ``c_0``: coeficiente del parámetro 0 (a_0).

    ``c_ind``: coeficiente del término independiente.
    """
    c_2 = sum(xi**2 for xi in xs)  # Suma de x^2
    c_1 = sum(xi for xi in xs)     # Suma de x
    c_0 = len(xs)                  # Número de puntos
    c_ind = sum(yi * xi**2 for xi, yi in zip(xs, ys))  # Suma de y * x^2

    return c_2, c_1, c_0, c_ind

def der_parcial_1_2(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 1 (a_1). La ecuación es de la forma:
    c_2 * a_2 + c_1 * a_1 + c_0 * a_0 = c_ind.

    ## Parameters

    ``xs``: lista de valores de x.

    ``ys``: lista de valores de y.

    ## Return

    ``c_2``: coeficiente del parámetro 2 (a_2).

    ``c_1``: coeficiente del parámetro 1 (a_1).

    ``c_0``: coeficiente del parámetro 0 (a_0).

    ``c_ind``: coeficiente del término independiente.
    """
    c_2 = sum(xi**3 for xi in xs)  # Suma de x^3
    c_1 = sum(xi**2 for xi in xs)  # Suma de x^2
    c_0 = sum(xi for xi in xs)     # Suma de x
    c_ind = sum(yi * xi for xi, yi in zip(xs, ys))  # Suma de y * x

    return c_2, c_1, c_0, c_ind

def der_parcial_0_2(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 0 (a_0). La ecuación es de la forma:
    c_2 * a_2 + c_1 * a_1 + c_0 * a_0 = c_ind.

    ## Parameters

    ``xs``: lista de valores de x.

    ``ys``: lista de valores de y.

    ## Return

    ``c_2``: coeficiente del parámetro 2 (a_2).

    ``c_1``: coeficiente del parámetro 1 (a_1).

    ``c_0``: coeficiente del parámetro 0 (a_0).

    ``c_ind``: coeficiente del término independiente.
    """
    c_2 = sum(xi**4 for xi in xs)  # Suma de x^4
    c_1 = sum(xi**3 for xi in xs)  # Suma de x^3
    c_0 = sum(xi**2 for xi in xs)  # Suma de x^2
    c_ind = sum(yi for yi in ys)   # Suma de y

    return c_2, c_1, c_0, c_ind
```
